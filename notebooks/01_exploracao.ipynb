```python
import sys
from pathlib import Path

sys.path.append(str(Path().resolve().parent))
import pandas as pd
from config.settings import RAW_DATA_DIR
```

Adicionando a raiz do projeto ao notebook

Tentando ler os arquivos com pd.read(), percebe-se um erro que diz que esperava-se 3 campos em uma linha e encontrou-se mais campos, provável erro de separador, para resolver esse erro, colocaremos None no campo sep, para garantir a letura tambem mudaremos o encoding padrão para latin1, para que o interpretador considere acentos. E caso o registro estaja muito quebrado, usaremos skip para pula-lo

```python
dfs = []
for file in list(RAW_DATA_DIR.glob('*.csv')):
    for enc in ('utf-8-sig', 'latin1'):
        try:
            df = pd.read_csv(
                file,
                sep=None,
                engine='python',
                encoding=enc,
                on_bad_lines='skip'
            )
            break
        except UnicodeDecodeError:
            continue
    #Padronizando os nomes das colunas
    df.columns = (
        df.columns              #É importante a utilização do atributo str, por que estamos tratando de strings, caso a leitura esbarre em um outro formato, pode ocasionar em erro
        .str.strip()
        .str.lower()
        .str.replace(' ', '_')
    )

    dfs.append(df)
```

```python
colunas = {}
for df in dfs:
    for coluna in df.columns:
        if coluna in colunas:
            colunas[coluna] +=1
        else:
            colunas[coluna] = 1

for chave, valor in colunas.items():
    print(f'{chave}: {valor}')
```

Podemos perceber que de todas as colunas, a coluna de região possui nomenclaturas diferentes, precisamos padronizar isso, para ajustar, vou remontar o carregamento dos csv
